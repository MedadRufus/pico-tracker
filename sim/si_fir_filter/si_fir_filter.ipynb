# Let's plot some filters 

```python
# Reflects an array of coefficients about the last element                                                                                                                                                                                                     
def sym_fir_coeff(coeff):
    for i in range(len(coeff)-2, -1, -1):
        coeff.append(coeff[i])

    return coeff

# Normalises the si's 8-bit coefficients to 0 - 1                                                                                                                                                                                                              
def si_normalise_coeff(coeff):
    return [x/0x300 for x in coeff] # 0x300 seems to be the value required to get 0 gain at dc.. Not sure about this: Test!
```

```python
# Default si coefficients                                                                                                                                                                                                                                      
si_default_fir_coeff = [0x1, 0x3, 0x8, 0x11, 0x21, 0x36, 0x4d, 0x60, 0x67]
si_fir_coeff = si_normalise_coeff(sym_fir_coeff(si_default_fir_coeff))
print si_fir_coeff
```

```python
%matplotlib inline
```

```python
from scipy import signal
from pylab import *
from __future__ import division
import matplotlib.pyplot as plt
```

```python
w,h=signal.freqz(si_fir_coeff,1) # Compute impulse response                                                                                                                                                                                                       

plt.plot(w/pi,20*log10(abs(h)))
plt.grid()
```

So this a Gausian Filter with B\*T=0.5. Nice

# Generate some test aprs data

```python
samp_rate = 60*1000
```

```python
def create_sine(frequency, amplitude=1.0, bits=1, bitrate=1200, sr=samp_rate):
        global phase
        
        w = 2.0 * pi * frequency
        t = arange(0, (bits / bitrate), bits / sr)
        
        wave = sin(w * t + phase)
        phase += (w * bits / bitrate)
        
        return wave
    
def random_data(length=200):
        return [0 if x < 0.5 else 1 for x in rand(length)]
```

```python
def get_afsk(len_afsk=25, sq=False, sr=samp_rate, dl=1, dh=1):
    global phase
    phase = pi/2
    
    if sq: # Square Wave
        afsk = concatenate([[-dl if x < 0 else dl for x in create_sine(1200, sr=sr)] if x 
                            else [-dh if x < 0 else dh for x in create_sine(2200, sr=sr)]
                            for x in random_data(len_afsk)])
    
    else:        
        afsk = concatenate([create_sine(1200, sr=sr) * dl if x 
                           else create_sine(2200, sr=sr) * dh for x in random_data(len_afsk)])
    
    
    
    return afsk
```

```python
plt.plot(get_afsk(len_afsk=12))
```

```python
plt.plot(get_afsk(len_afsk=10, sq=True, dh=3/1.6)) # test out the different deviations for high/low tone
```

```python
# Plots a set of afsk spectrums

def plot_afsk_spectrum(afsks, num="", sr=samp_rate, p_avg=True, p_max=True):
    S = sr/(1200*15)
    len_afsk = 25

    n_afsk = (len_afsk/1200)*samp_rate # Total number of samples
    
    win = blackman(len(afsks[0]))
    
    ffts = [fft(afsk*win) for afsk in afsks]

    yf_average = average(ffts, axis=0)
    yf_maximum = amax(ffts, axis=0)

    xf = np.linspace(0.0, 0.5*sr/S, n_afsk/(2*S))
    
    if p_avg:
        avg, = plt.semilogy(xf, 2* 2.0 * np.abs(yf_average[0:n_afsk/(2*S)]), label='Avg'+num)
    if p_max:
        maxi, = plt.semilogy(xf, 2.0 * np.abs(yf_maximum[0:n_afsk/(2*S)]), label='Max'+num)
    
    if p_avg and p_max:
        plt.legend(handles=[avg,maxi])
        
    plt.grid()
    
def plot_afsk_spectrum2(afsks1, afsks2, p_avg=True, p_max=True):
    plot_afsk_spectrum(afsks1, "1", p_avg=p_avg, p_max=p_max)
    plot_afsk_spectrum(afsks2, "2", p_avg=p_avg, p_max=p_max)
    plt.grid()
```

We can see that while on average the energy is at 1.2kHz and 2.2kHz there's actually things going to up to 3kHz or so

```python
afsks = [get_afsk(len_afsk) for x in range(100)]

plot_afsk_spectrum(afsks)
```

```python
afsks = [get_afsk(len_afsk, sq = False) for x in range(100)]
afsks_sq = [get_afsk(len_afsk, sq = True) for x in range(100)]

plot_afsk_spectrum2(afsks, afsks_sq)
```

# What happens when we add a filter?
Let's tune the deviation ratios too

Start by defining the tx data rate `DATA_RATE` and oversampling ratio `TXOSR` 

```python
tx_data_rate = 3000
tx_osr = 10

# Our filter samples at the product of these two values
filter_samp_rate = tx_data_rate * tx_osr
```

```python
# Deviation for high tone
devl = 1800     # Adjust to give 1600hz deviation on the squarewave input
devh = 4800     # Adjust to give 3000hz deviation on the squareware input
```

```python
# Which fir filter to use
fir = si_fir_coeff
```

```python
afsks = [get_afsk(len_afsk, sq = False, sr=filter_samp_rate, dl=devl, dh=devh) for x in range(100)]
afsks_filtered = [convolve(afsk, fir, mode='valid') for afsk in afsks] 

afsk_time = get_afsk(len_afsk=12, sq = False, sr=filter_samp_rate, dl=devl, dh=devh)
plt.plot(convolve(afsk_time, fir, mode='valid'))
```
