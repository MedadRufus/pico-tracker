Simplify

```python
# 
# from http://mappinghacks.com/code/dp.py.txt
#

# pure-Python Douglas-Peucker line simplification/generalization
#
# this code was written by Schuyler Erle <schuyler@nocat.net> and is
#   made available in the public domain.
#
# the code was ported from a freely-licensed example at
#   http://www.3dsoftware.com/Cartography/Programming/PolyLineReduction/
#
# the original page is no longer available, but is mirrored at
#   http://www.mappinghacks.com/code/PolyLineReduction/

"""

>>> line = [(0,0),(1,0),(2,0),(2,1),(2,2),(1,2),(0,2),(0,1),(0,0)]
>>> simplify_points(line, 1.0)
[(0, 0), (2, 0), (2, 2), (0, 2), (0, 0)]

>>> line = [(0,0),(0.5,0.5),(1,0),(1.25,-0.25),(1.5,.5)]
>>> simplify_points(line, 0.25)
[(0, 0), (0.5, 0.5), (1.25, -0.25), (1.5, 0.5)]

"""

import math
   
def simplify_points (pts, tolerance): 
    anchor  = 0
    floater = len(pts) - 1
    stack   = []
    keep    = set()

    stack.append((anchor, floater))  
    while stack:
        anchor, floater = stack.pop()
      
        # initialize line segment
        if pts[floater] != pts[anchor]:
            anchorX = float(pts[floater][0] - pts[anchor][0])
            anchorY = float(pts[floater][1] - pts[anchor][1])
            seg_len = math.sqrt(anchorX ** 2 + anchorY ** 2)
            # get the unit vector
            anchorX /= seg_len
            anchorY /= seg_len
        else:
            anchorX = anchorY = seg_len = 0.0
    
        # inner loop:
        max_dist = 0.0
        farthest = anchor + 1
        for i in range(anchor + 1, floater):
            dist_to_seg = 0.0
            # compare to anchor
            vecX = float(pts[i][0] - pts[anchor][0])
            vecY = float(pts[i][1] - pts[anchor][1])
            seg_len = math.sqrt( vecX ** 2 + vecY ** 2 )
            # dot product:
            proj = vecX * anchorX + vecY * anchorY
            if proj < 0.0:
                dist_to_seg = seg_len
            else: 
                # compare to floater
                vecX = float(pts[i][0] - pts[floater][0])
                vecY = float(pts[i][1] - pts[floater][1])
                seg_len = math.sqrt( vecX ** 2 + vecY ** 2 )
                # dot product:
                proj = vecX * (-anchorX) + vecY * (-anchorY)
                if proj < 0.0:
                    dist_to_seg = seg_len
                else:  # calculate perpendicular distance to line (pythagorean theorem):
                    dist_to_seg = math.sqrt(abs(seg_len ** 2 - proj ** 2))
                if max_dist < dist_to_seg:
                    max_dist = dist_to_seg
                    farthest = i

        if max_dist <= tolerance: # use line segment
            keep.add(anchor)
            keep.add(floater)
        else:
            stack.append((anchor, farthest))
            stack.append((farthest, floater))

    keep = list(keep)
    keep.sort()
    return [pts[i] for i in keep]
```

```python
def prepare_simple_grids(grids, tolerance=2*8000): # Approx 8km
    simple = [simplify_points(points, tolerance) for points in grids]
    return [s for s in simple if len(s) > 2]
```

```python
# We'll need a utility function for splitting outlines that cross the antiprime meridian
def fix_antiprime_outlines(outlines):
    new_outlines = []
    
    for outline in outlines:
        
        last_lon = 0
        last_lat = 0
        a_outline = []
        b_outline = []
        b_side = False
        
        for point in outline:
            
            # Possibly change sides
            if abs(last_lon - point[0]) > 180: # If we just jumped more than 180˚ of longitude
                interpolate_lat = point[1]
                
                if b_side:
                    b_outline.append((180 if last_lon > 0 else -180, interpolate_lat))
                else:
                    a_outline.append((180 if last_lon > 0 else -180, interpolate_lat))
                    
                b_side = not b_side
                print "Swap!"
                
                if b_side:
                    b_outline.append((180 if point[0] > 0 else -180, interpolate_lat))
                else:
                    a_outline.append((180 if point[0] > 0 else -180, interpolate_lat))
                
            if b_side:
                b_outline.append(point)
            else:
                a_outline.append(point)
                
            last_lon = point[0]
            last_lat = point[1]
                
        new_outlines.append(a_outline)
        new_outlines.append(b_outline)
            
            
    return new_outlines
```

```python
# We'll use pyclipper to offset / union / difference our grids
import pyclipper
```

```python
# Define functions for translating to/from grids
import mpl_toolkits.basemap.pyproj as pyproj # Import the pyproj module
 
wgs84=pyproj.Proj(proj='merc', ellps='WGS84') # LatLon with WGS84 datum

# lon/lat to x/y grid
def lonlat_to_ea_grid(outlines):
    return [[wgs84(point[0], point[1]) for point in outline] for outline in outlines]
        
# x/y grid to lon/lat
def ea_grid_to_lonlat(outlines):
    return [[wgs84(point[0], point[1], inverse=True) for point in outline] for outline in outlines]
```

```python
# Now expand and union
all_countries_shifted_grids = []

# Offset the grids by ~600km
for grid in all_countries_grids:
    pc = pyclipper.PyclipperOffset()    
        
    pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
    shifted_grid = [(point[0], point[1]) for point in pc.Execute(600 *2*1000)[0]]
        
    all_countries_shifted_grids.append(shifted_grid)    
    
# Union
pc = pyclipper.Pyclipper()
pc.AddPaths(all_countries_shifted_grids, pyclipper.PT_SUBJECT, True)
    
unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
all_countries_shifted_grids = [[(point[0], point[1]) for point in union] for union in unions]
```

Export to header file

```python
def file_header_telemetry(f):
    print >>f, "/**"
    print >>f, " * Autogenerated telemetry zone structures. See sim/geofence"
    print >>f, " */"
    print >>f, ""
    
def header_h_telemetry(header, no_telem_outlines):
    print >>header, "#ifndef GEOFENCE_TELEMETRY_H"
    print >>header, "#define GEOFENCE_TELEMETRY_H"
    print >>header, ""
    print >>header, ""
    print >>header, "#include <stdbool.h>"
    print >>header, ""
    print >>header, "#include \"samd20.h\""
    print >>header, ""
    print >>header, "const int32_t* no_telem_outlines[{0}];".format(len(no_telem_outlines))
    print >>header, "const uint32_t no_telem_outline_lengths[{0}];".format(len(no_telem_outlines))
    print >>header, ""
    print >>header, "#endif /* GEOFENCE_APRS_H */"
```

```python
def print_telemetry(f, no_telem_outlines):
    
    # Print each array of points
    for o in range(len(no_telem_outlines)): # foreach of the countries outlines
    
        outline = no_telem_outlines[o]
    
        print >>f, "const int32_t no_telemetry_outline_{0}[] = {{".format(o+1)
        for point in outline:
            # We express locations in µdegrees
            ulon = int(round(point[0]*1e6))
            ulat = int(round(point[1]*1e6))
        
            print >>f, "  {: 10d}, {: 9d},".format(ulon, ulat)
        print >>f, "};"

            
    print >>f, ""
    
    # Print an array of outlines
    print >>f, "const int32_t* no_telem_outlines[] = {"
    for o in range(len(no_telem_outlines)): # foreach of the outlines
         print >>f, "  no_telemetry_outline_{0},".format(o+1)
    print >>f, "};"
    
    # Print an array of their lengths
    print >>f, "const uint32_t no_telem_outline_lengths[] = {"
    for outline in no_telem_outlines: # foreach of the outlines
         print >>f, "  {0},".format(len(outline))
    print >>f, "};"
    print >>f, ""
    print >>f, ""
```

```python
def write_telemetry_zones():
    # header
    header = open('../../firmware/inc/geofence_telemetry.h','w')

    file_header_telemetry(header)
    header_h_telemetry(header, no_telem_outlines)

    # source
    source = open('../../firmware/src/geofence_telemetry.c', 'w')

    file_header_telemetry(source)
    print >>source, ""
    print >>source, "#include <stdbool.h>"
    print >>source, ""
    print >>source, "#include \"samd20.h\""
    print >>source, "#include \"geofence_telemetry.h\""
    print >>source, ""
    print >>source, ""
    print >>source, "/* Longitude,  Latitude */"
    print >>source
    print_telemetry(source, no_telem_outlines)

    header.close()
    source.close()
```

```python
write_telemetry_zones()
```
