```python
def prepare_simple_grids(grids, tolerance=5*1000): # 5km tolerance
    simple = [simplify_points(points, tolerance) for points in grids]
    return [s for s in simple if len(s) > 2]
```

##Build up a grid of mid-ocean zones

```python
# Union the grids of all countries
all_countries_grids = []
    
# Gather the simple grids for all the countries
for c in countries:
    all_countries_grids = all_countries_grids + c['simple_grids']
```

```python
# Now expand and union
all_countries_shifted_grids = []

# Offset the grids by 600km
for grid in all_countries_grids:
    pc = pyclipper.PyclipperOffset()    
        
    pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
    shifted_grid = [(point[0], point[1]) for point in pc.Execute(600*1000)[0]]
        
    all_countries_shifted_grids.append(shifted_grid)    
    
# Union
pc = pyclipper.Pyclipper()
pc.AddPaths(all_countries_shifted_grids, pyclipper.PT_SUBJECT, True)
    
unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
all_countries_shifted_grids = [[(point[0], point[1]) for point in union] for union in unions]
```

##Build up no telemetry grids

```python
no_telem_grids = []
    
# Gather the simple grids for all the countries in this zone
for c in countries:
    if c['isocode'] in telemetry_tx_exclude:
        no_telem_grids = no_telem_grids + c['shifted_grids']
    
# Union
pc = pyclipper.Pyclipper()
pc.AddPaths(no_telem_grids, pyclipper.PT_SUBJECT, True)
    
unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
combined_grids = [[(point[0], point[1]) for point in union] for union in unions]
    
combined_grids = prepare_simple_grids(combined_grids, tolerance=10*1000) # 10km tolerance
print len(combined_grids)
print sum([len(g) for g in combined_grids])
    
no_telem_grids = combined_grids
no_telem_outlines = [o for o in fix_antiprime_outlines(ea_grid_to_lonlat(no_telem_grids)) if o]
print
print len(no_telem_outlines)
```
