```python
import xml.etree.ElementTree as et
import os

root = et.parse('./countries_world.kml').getroot()
```

Let's parse this kml file.

First define some functions for extracting various fields

```python
import re

# name
def c_name(e):
    return e.text
# description
def c_description(e):
    return e.text
# iso country code
def c_isocode(name, description):
    matchobj = re.search(r'ISO_A2=([A-Z]{2})', description.text)
    
    if matchobj: # Valid iso country code
        return matchobj.groups()[0]
    else: # No country code.. return "x?"
        return name.text[0]+'?'
    
# centre
def c_centre(e):
    lat = e.find('{http://www.opengis.net/kml/2.2}latitude').text
    lon = e.find('{http://www.opengis.net/kml/2.2}longitude').text
    
    return (float(lon),float(lat))
# colour
def c_colour(e):
    return e.find('{http://www.opengis.net/kml/2.2}PolyStyle').find('{http://www.opengis.net/kml/2.2}color').text
# outline
def c_outline(e):
    # extract the coords strings from the xml
    coord_sets = e.findall('.{http://www.opengis.net/kml/2.2}Polygon//{http://www.opengis.net/kml/2.2}coordinates')
    
    coord_sets = [[(float(c.split(',')[0]),float(c.split(',')[1])) for c in coords.text.split()] for coords in coord_sets]
    
    return coord_sets
```

```python
countries = root.findall('.//{http://www.opengis.net/kml/2.2}Placemark')
countries = [{
    'name':         c_name(c.find('{http://www.opengis.net/kml/2.2}name')),
    'description':  c_description(c.find('{http://www.opengis.net/kml/2.2}description')),
    'isocode':      c_isocode(c.find('{http://www.opengis.net/kml/2.2}name'), c.find('{http://www.opengis.net/kml/2.2}description')),
    'centre':       c_centre(c.find('{http://www.opengis.net/kml/2.2}LookAt')),
    'colour':       c_colour(c.find('{http://www.opengis.net/kml/2.2}Style')),
    'outlines':     c_outline(c.find('{http://www.opengis.net/kml/2.2}MultiGeometry')),
} for c in countries]
```

#### C header

```python
def file_header(f):
    print >>f, "/**"
    print >>f, " * Autogenerated country structures. See sim/geofence"
    print >>f, " */"
    print >>f, ""
    
def header_h(h):
    print >>header, "#ifndef GEOFENCE_H"
    print >>header, "#define GEOFENCE_H"
    print >>header, ""
    print >>header, ""
    print >>header, "#include \"samd20.h\""
    print >>header, ""
    print >>header, ""
    print >>header, "/**"
    print >>header, " * struct representing all we need to know about a country"
    print >>header, " */"
    print >>header, "struct country_t {"
    print >>header, "  int32_t* outline;"
    print >>header, "  char* name, isocode;"
    print >>header, "  bool data_tx, aprs_tx;"
    print >>header, "  uint32_t aprs_freq, aprs_dev;"
    print >>header, "};"
    print >>header, ""
    print >>header, ""
    print >>header, "struct country_t countries[];"
    print >>header, ""
    print >>header, "#endif /* GEOFENCE_H */"
```

We print the country outlines into an array `int32_t`

```python
def print_outline(f, country):
    
    print >>f, u"/* {name:->24} */".format(**country).encode('utf8')
    
    for o in range(len(country['outlines'])): # foreach of the countries outlines
    
        outline = country['outlines'][o]
    
        print >>f, "int32_t {0}_{1}[] = {{".format(country['isocode'].lower() + "_outline", o+1)
        for point in outline:
            # We express locations in µdegrees
            ulon = int(round(point[0]*1e6))
            ulat = int(round(point[1]*1e6))
        
            print >>f, "  {:= 010d}, {:= 09d},".format(ulon, ulat)
        print >>f, "};"
        
    print >>f, ""
```

Print out a structure containing the various values for each country and a pointer to the outline

```python
from __future__ import unicode_literals

def print_country(f, country):
    country['outline_n'] = country['isocode'].lower() + "_outlines"
    country['name'] = unicode(country['name'])
    
    print >>f, ""
    print >>f, u"                                              \
    /* {name:->24} */".format(**country).encode('utf8')
    print >>f, u"  {{ .outlines = {outline_n}, .isocode = \"{isocode}\", .data_tx = {data_tx}, .aprs_tx = {aprs_tx},\n\
    .aprs_freq = {aprs_freq}, .aprs_dev = {aprs_dev}, .name = \"{name}\" }},".format(**country).encode('utf8')
    
def print_countries(f, countries):
    print >>f, "struct country_t countries[] = {"
    [print_country(f, c) for c in countries]
    print >>f, "};"
```

Let's define the point in poly function

```python
#  Point in poly
#
#  int    polyCorners  =  how many corners the polygon has (no repeats)
#  float  polyX[]      =  horizontal coordinates of corners
#  float  polyY[]      =  vertical coordinates of corners
#  float  x, y         =  point to be tested
#
#  The function will return YES if the point x,y is inside the polygon, or
#  NO if it is not.  If the point is exactly on the edge of the polygon,
#  then the function may return YES or NO.
#
#  Note that division by zero is avoided because the division is protected
#  by the "if" clause which surrounds it.

def point_in_polygon(outline, point):

    nodes = 0
    y = point[0]
    x = point[1]

    for i in range(len(outline)):   # for (i=0; i<polyCorners; i++) {

        if ((outline[i][0] < y and outline[i-1][0] >= y) or (outline[i-1][0] < y and outline[i][0] >= y))\
            and (outline[i][1] <= x or outline[i-1][1] <= x):

            #if (polyX[i]+(y-polyY[i])/(polyY[j]-polyY[i])*(polyX[j]-polyX[i])<x)
            if (outline[i][1] + (y-outline[i][0]) / (outline[i-1][0]-outline[i][0]) * (outline[i-1][1]-outline[i][1]) < x):
                nodes = nodes + 1
    
    # Odd means we're in the polygon
    return True if (nodes % 2 == 1) else False
```

```python
def point_in_country(country, point):
    for outline in country['outlines']:
        if point_in_polygon(outline, point):
            return True
    
    return False
```
