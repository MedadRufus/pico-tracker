First define some functions for extracting various fields

```python
import re

# name
def c_name(e):
    return e.text
# description
def c_description(e):
    return e.text
# iso country code
def c_isocode(name, description):
    matchobj = re.search(r'ISO_A2=([A-Z]{2})', description.text)
    
    if matchobj: # Valid iso country code
        return matchobj.groups()[0]
    else: # No country code.. return "x0"
        return name.text[0]+'0'
    
# centre
def c_centre(e):
    lat = e.find('{http://www.opengis.net/kml/2.2}latitude').text
    lon = e.find('{http://www.opengis.net/kml/2.2}longitude').text
    
    return (float(lon),float(lat))
# colour
def c_colour(e):
    return e.find('{http://www.opengis.net/kml/2.2}PolyStyle').find('{http://www.opengis.net/kml/2.2}color').text
# outline
def c_outline(e):
    # extract the coords strings from the xml
    coord_sets = e.findall('.{http://www.opengis.net/kml/2.2}Polygon//{http://www.opengis.net/kml/2.2}coordinates')
    
    coord_sets = [[(float(c.split(',')[0]),float(c.split(',')[1])) for c in coords.text.split()] for coords in coord_sets]
    
    return coord_sets
```

```python
countries = root.findall('.//{http://www.opengis.net/kml/2.2}Placemark')
countries = [{
    'name':         c_name(c.find('{http://www.opengis.net/kml/2.2}name')),
    'description':  c_description(c.find('{http://www.opengis.net/kml/2.2}description')),
    'isocode':      c_isocode(c.find('{http://www.opengis.net/kml/2.2}name'), c.find('{http://www.opengis.net/kml/2.2}description')),
    'centre':       c_centre(c.find('{http://www.opengis.net/kml/2.2}LookAt')),
    'colour':       c_colour(c.find('{http://www.opengis.net/kml/2.2}Style')),
    'outlines':     c_outline(c.find('{http://www.opengis.net/kml/2.2}MultiGeometry')),
} for c in countries]
```

#### C header

```python
def file_header(f):
    print >>f, "/**"
    print >>f, " * Autogenerated country structures. See sim/geofence"
    print >>f, " */"
    print >>f, ""
    
def header_h(header, countries):
    print >>header, "#ifndef GEOFENCE_COUNTRIES_H"
    print >>header, "#define GEOFENCE_COUNTRIES_H"
    print >>header, ""
    print >>header, ""
    print >>header, "#include <stdbool.h>"
    print >>header, ""
    print >>header, "#include \"samd20.h\""
    print >>header, ""
    print >>header, ""
    print >>header, "/**"
    print >>header, " * struct representing all we need to know about a country"
    print >>header, " */"
    print >>header, "struct country_t {"
    print >>header, "  const int32_t** outlines;"
    print >>header, "  const int32_t outline_count;"
    print >>header, "  const uint32_t* outline_lengths;"
    print >>header, "  char* name;"
    print >>header, "  char* isocode;"
    print >>header, "  bool data_tx, aprs_tx;"
    print >>header, "};"
    print >>header, ""
    print >>header, ""
    print >>header, "const struct country_t countries[{0}];".format(len(countries))
    print >>header, ""
    print >>header, "#endif /* GEOFENCE_COUNTRIES_H */"
```

We print the country outlines into an array `int32_t`

```python
def print_outline(f, country):
    
    print >>f, u"/* {name:->24} */".format(**country).encode('utf8')
    
    # Print each array of points
    for o in range(len(country['simple_outlines'])): # foreach of the countries outlines
    
        outline = country['simple_outlines'][o]
    
        print >>f, "const int32_t {0}_{1}[] = {{".format(country['isocode'].lower() + "_outline", o+1)
        for point in outline:
            # We express locations in µdegrees
            ulon = int(round(point[0]*1e6))
            ulat = int(round(point[1]*1e6))
        
            print >>f, "  {: 10d}, {: 9d},".format(ulon, ulat)
        print >>f, "};"
        
    print >>f, ""
    
    # Print an array of outlines
    print >>f, "const int32_t* {0}_outlines[] = {{".format(country['isocode'].lower())
    for o in range(len(country['simple_outlines'])): # foreach of the countries outlines
         print >>f, "  {0}_{1},".format(country['isocode'].lower() + "_outline", o+1)
    print >>f, "};"
    
    # Print an array of their lengths
    print >>f, "const uint32_t {0}_outline_lengths[] = {{".format(country['isocode'].lower())
    for outline in country['simple_outlines']: # foreach of the countries outlines
         print >>f, "  {0},".format(len(outline))
    print >>f, "};"
    print >>f, ""
```

Print out a structure containing the various values for each country and a pointer to the outline

```python
from __future__ import unicode_literals

def print_country(f, country):
    country['outline_n'] = country['isocode'].lower() + "_outlines"
    country['outline_c'] = len(country['simple_outlines'])
    country['outline_l'] = country['isocode'].lower() + "_outline_lengths"
    country['name'] = unicode(country['name'])
    
    print >>f, ""
    print >>f, u"                                              \
    /* {name:->24} */".format(**country).encode('utf8')
    print >>f, u"  {{ .outlines = {outline_n}, .outline_count = {outline_c}, .outline_lengths = {outline_l}, .isocode = \"{isocode}\",\n\
    .data_tx = {data_tx}, .aprs_tx = {aprs_tx}, .name = \"{name}\" }},".format(**country).encode('utf8')
    
def print_countries(f, countries):
    print >>f, "const struct country_t countries[] = {"
    [print_country(f, c) for c in countries]
    print >>f, "};"
```

Let's define the point in poly function

```python
#  Point in poly
#
#  int    polyCorners  =  how many corners the polygon has (no repeats)
#  float  polyX[]      =  horizontal coordinates of corners
#  float  polyY[]      =  vertical coordinates of corners
#  float  x, y         =  point to be tested
#
#  The function will return YES if the point x,y is inside the polygon, or
#  NO if it is not.  If the point is exactly on the edge of the polygon,
#  then the function may return YES or NO.
#
#  Note that division by zero is avoided because the division is protected
#  by the "if" clause which surrounds it.

def point_in_polygon(outline, point):

    nodes = 0
    y = point[0]
    x = point[1]

    for i in range(len(outline)):   # for (i=0; i<polyCorners; i++) {

        if ((outline[i][0] < y and outline[i-1][0] >= y) or (outline[i-1][0] < y and outline[i][0] >= y))\
            and (outline[i][1] <= x or outline[i-1][1] <= x):

            #if (polyX[i]+(y-polyY[i])/(polyY[j]-polyY[i])*(polyX[j]-polyX[i])<x)
            if (outline[i][1] + (y-outline[i][0]) / (outline[i-1][0]-outline[i][0]) * (outline[i-1][1]-outline[i][1]) < x):
                nodes = nodes + 1
    
    # Odd means we're in the polygon
    return True if (nodes % 2 == 1) else False
```

```python
def point_in_country(country, point):
    for outline in country['outlines']:
        if point_in_polygon(outline, point):
            return True
    
    return False
```

We want to roject them onto an equal-area, integer grid and we can then use pyclipper to offset them

```python
# Define functions for translating to/from grids
import mpl_toolkits.basemap.pyproj as pyproj # Import the pyproj module
 
wgs84=pyproj.Proj(proj='merc', ellps='WGS84') # LatLon with WGS84 datum

# lon/lat to x/y grid
def lonlat_to_ea_grid(outlines):
    return [[wgs84(point[0], point[1]) for point in outline] for outline in outlines]
        
# x/y grid to lon/lat
def ea_grid_to_lonlat(outlines):
    return [[wgs84(point[0], point[1], inverse=True) for point in outline] for outline in outlines]
```

```python
# Convert all countries to grids
for c in countries:
    c['grids'] = lonlat_to_ea_grid(c['outlines'])
```

```python
# Utility functions
def all_grids_all_countries_except(isocode = 'GB'):
    grids = []
    
    for c in countries:
        if c['isocode'] != isocode:
            for g in c['grids']:
                grids.append(g)
            
    return grids
```

```python
def prepare_shifted_grids(grids, isocode):
    shifted_grids = []

    # Offset the grids by ~25km (terratorial waters)
    for grid in grids:
        pc = pyclipper.PyclipperOffset()    
        
        pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
        shifted_grid = [(point[0], point[1]) for point in pc.Execute(25 *2*1000)[0]]
        
        shifted_grids.append(shifted_grid)    
    
    # Union
    pc = pyclipper.Pyclipper()
    pc.AddPaths(shifted_grids, pyclipper.PT_SUBJECT, True)
    
    unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
    shifted_grids = [[(point[0], point[1]) for point in union] for union in unions]
        
    # Difference with other countries
    other_grids = all_grids_all_countries_except(isocode)
    pc = pyclipper.Pyclipper()
    pc.AddPaths(shifted_grids, pyclipper.PT_SUBJECT, True)
    pc.AddPaths(other_grids, pyclipper.PT_CLIP, True)
    
    differences = pc.Execute(pyclipper.CT_DIFFERENCE, pyclipper.PFT_EVENODD, pyclipper.PFT_EVENODD)
    shifted_grids = [[(point[0], point[1]) for point in difference] for difference in differences]
    #
    # This isn't quite ideal but it works
    #
    
    return shifted_grids

# Test for GB
gb_shifted_grids = prepare_shifted_grids(gb['grids'], "GB")

# Back to lat/lon
gb_shifted = ea_grid_to_lonlat(gb_shifted_grids)

draw_gb(outlines=gb_shifted)
```

#### Simply line

```python
# 
# from http://mappinghacks.com/code/dp.py.txt
#

# pure-Python Douglas-Peucker line simplification/generalization
#
# this code was written by Schuyler Erle <schuyler@nocat.net> and is
#   made available in the public domain.
#
# the code was ported from a freely-licensed example at
#   http://www.3dsoftware.com/Cartography/Programming/PolyLineReduction/
#
# the original page is no longer available, but is mirrored at
#   http://www.mappinghacks.com/code/PolyLineReduction/

"""

>>> line = [(0,0),(1,0),(2,0),(2,1),(2,2),(1,2),(0,2),(0,1),(0,0)]
>>> simplify_points(line, 1.0)
[(0, 0), (2, 0), (2, 2), (0, 2), (0, 0)]

>>> line = [(0,0),(0.5,0.5),(1,0),(1.25,-0.25),(1.5,.5)]
>>> simplify_points(line, 0.25)
[(0, 0), (0.5, 0.5), (1.25, -0.25), (1.5, 0.5)]

"""

import math
   
def simplify_points (pts, tolerance): 
    anchor  = 0
    floater = len(pts) - 1
    stack   = []
    keep    = set()

    stack.append((anchor, floater))  
    while stack:
        anchor, floater = stack.pop()
      
        # initialize line segment
        if pts[floater] != pts[anchor]:
            anchorX = float(pts[floater][0] - pts[anchor][0])
            anchorY = float(pts[floater][1] - pts[anchor][1])
            seg_len = math.sqrt(anchorX ** 2 + anchorY ** 2)
            # get the unit vector
            anchorX /= seg_len
            anchorY /= seg_len
        else:
            anchorX = anchorY = seg_len = 0.0
    
        # inner loop:
        max_dist = 0.0
        farthest = anchor + 1
        for i in range(anchor + 1, floater):
            dist_to_seg = 0.0
            # compare to anchor
            vecX = float(pts[i][0] - pts[anchor][0])
            vecY = float(pts[i][1] - pts[anchor][1])
            seg_len = math.sqrt( vecX ** 2 + vecY ** 2 )
            # dot product:
            proj = vecX * anchorX + vecY * anchorY
            if proj < 0.0:
                dist_to_seg = seg_len
            else: 
                # compare to floater
                vecX = float(pts[i][0] - pts[floater][0])
                vecY = float(pts[i][1] - pts[floater][1])
                seg_len = math.sqrt( vecX ** 2 + vecY ** 2 )
                # dot product:
                proj = vecX * (-anchorX) + vecY * (-anchorY)
                if proj < 0.0:
                    dist_to_seg = seg_len
                else:  # calculate perpendicular distance to line (pythagorean theorem):
                    dist_to_seg = math.sqrt(abs(seg_len ** 2 - proj ** 2))
                if max_dist < dist_to_seg:
                    max_dist = dist_to_seg
                    farthest = i

        if max_dist <= tolerance: # use line segment
            keep.add(anchor)
            keep.add(floater)
        else:
            stack.append((anchor, farthest))
            stack.append((farthest, floater))

    keep = list(keep)
    keep.sort()
    return [pts[i] for i in keep]
```

```python
def prepare_simple_grids(grids, tolerance=2*5000): # Approx 5km
    simple = [simplify_points(points, tolerance) for points in grids]
    return [s for s in simple if len(s) > 2]

# Test with GB
gb_simple_grids = prepare_simple_grids(gb_shifted_grids)

# Back to lat/lon
gb_simple = ea_grid_to_lonlat(gb_simple_grids)

print len(gb_simple[1])

#print gb_simple
draw_gb(outlines=gb_simple)
```

```python
# We'll need a utility function for splitting outlines that cross the antiprime meridian
def fix_antiprime_outlines(outlines):
    new_outlines = []
    
    for outline in outlines:
        
        east = False
        west = False
        for point in outline:
            if point[0] > 150: east = True
            if point[0] < -150: west = True
        
        if east and west: # This outline crosses the antiprime. Presume it doesn't cross the -30th meridian
            east_outline = []
            west_outline = []
            
            for point in outline: # Split the outline
                lon = point[0]
                east_outline.append((lon if lon > -30 else  180, point[1]))
                west_outline.append((lon if lon < -30 else -180, point[1]))
            
            east_outline_strip = []
            west_outline_strip = []
            
            for i in range(len(east_outline)): # Strip runs of points that lie on the meridian
                if not (east_outline[i-1][0] == 180 and east_outline[i][0] == 180 and east_outline[(i+1)%len(east_outline)][0] == 180):
                    east_outline_strip.append(east_outline[i])
                    
            for i in range(len(west_outline)):
                if not(west_outline[i-1][0] == -180 and west_outline[i][0] == -180 and west_outline[(i+1)%len(west_outline)][0] == -180):
                    west_outline_strip.append(west_outline[i])
                    
            #print "Remove", len(outline)
            #print "East", len(east_outline_strip)
            #print "West", len(west_outline_strip)
             
            # Swap in these outlines
            new_outlines.append(east_outline_strip)
            new_outlines.append(west_outline_strip)
        else:
            new_outlines.append(outline)
            
    return new_outlines
```

```python
from operator import itemgetter, attrgetter, methodcaller
#
# Source: http://aprsisce.wikidot.com/doc:frequencies, http://aprsisce.wdfiles.com/local--files/doc:frequencies/APRSVHFworldmap.png
#
# The following areas are on 144.8 and 3000Hz deviation. This is the default.
# Europe, Africa, Middle East, Russia, S Africa, Costa Rica, Israel, Lebanon, Senegal
#
aprs_zones = [ # ---- North American 144.39 and others ----
              (144390e3, 5e3, 14, 600, ["US", "CA"]), # North America
              (144390e3, 3e3, 6,  300, [ # Chile, Indonesia, Singapore, Malaysia, Mexico, Dominican Rep, Puerto Rico, Trinidad&T, Columbi
                                     "CL", "ID", "SG", "MY", "MX", "DO", "PR", "TT", "CO"]),
              
               # ---- Various exceptions -----
              (145175e3, 3e3, 13, 600, ["AU"]), # Austrailia, Tasmania
              (144575e3, 3e3, 14, 400, ["NZ"]), # New Zealand
              (144930e3, 3e3, 2,  200, ["AR", "UY", "PY"]), # Argentina, Uruguay, Paraguay
              (144620e3, 3e3, 8,   80, ["KR"]), # South Korea
              (144640e3, 3e3, 7,  300, ["CN", "HK"]), # China, Hong Kong, Taiwan???!
              (144660e3, 3e3, 2,  500, ["JP"]), # Japan (1200 baud band)
              (145525e3, 3e3, 10, 100, ["TH"]), # Thailand
              (145570e3, 3e3, 11, 250, ["BR"]), # Brazil
              (145010e3, 3e3, 12, 200, ["VE", "PA"]), # Venezuela, Panama
              
              # ---- No Radio ----
              (0, 0, 20, 0, aprs_tx_exclude),    # North Korea, Yemen
             ]

aprs_zones.sort(key=itemgetter(2), reverse=True)

others = []
for c in countries:
    has_aprsinfo = False
    for a in aprs_zones:
        if c['isocode'] in a[4]:
            has_aprsinfo = True
    if not has_aprsinfo:
        #print c['name'], c['isocode']
        others.append(c['isocode'])

# Append this as another zone
aprs_zones.append((144800e3, 3e3, 0, 0, others))

# Zone names
names = []
dictionary = {'A':'Alpha', 'B':'Bravo','C':'Charlie', 'D':'Delta', 'E':'Echo', 'F':'Foxtrot', 
              'G':'Golf', 'I':'India', 'J':'Juliet', 'K':'Kilo', 'L':'Lima', 'M':'Mike', 
              'N':'November', 'O':'Oscar', 'P':'Papa', 'Q':'Quebec', 'R':'Romeo', 'S':'Sierra', 
              'T':'Tango', 'U':'Uniform', 'V':'Victor', 'W':'Whiskey', 'X':'Xray', 'Y':'Yankee', 'Z':'Zulu'}
for letter, name in dictionary.iteritems():
    names.append(name)

# Zone colours
colours = []
for name, hex in matplotlib.colors.cnames.iteritems():
    colours.append(hex)
        
# Process Zones
aprs_zones_obj = []
for z in range(len(aprs_zones)):
    zone = {}
    zone['frequency'] = int(aprs_zones[z][0])
    zone['deviation'] = int(aprs_zones[z][1])
    zone['isocodes'] = aprs_zones[z][4]
    zone['colour'] = colours[z]
    zone['name'] = names[z]
    zone['priority'] = aprs_zones[z][2]
    zone['expand'] = aprs_zones[z][3]
    
    
    aprs_zones_obj.append(zone)
        
print aprs_zones_obj
```

#### Now let's combine the terratorial waters of these coutries to make outlines for each zone

```python
for zone in aprs_zones_obj:
    grids = []
    
    # Gather the simple grids for all the countries in this zone
    for c in countries:
        if c['isocode'] in zone['isocodes']:
            grids = grids + c['shifted_grids']
    
    # Union
    pc = pyclipper.Pyclipper()
    pc.AddPaths(grids, pyclipper.PT_SUBJECT, True)
    
    unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
    combined_grids = [[(point[0], point[1]) for point in union] for union in unions]
    
    combined_grids = prepare_simple_grids(combined_grids, tolerance=2000)
    #print len(combined_grids)
    
    zone['grids'] = combined_grids
    zone['outlines'] = fix_antiprime_outlines(ea_grid_to_lonlat(zone['grids']))
```

#### Now we want to expand this to cover the whole globe....

We won't expand the 144.8 polygon and will discard it eventually

```python
for zone in aprs_zones_obj:
    grids = zone['grids']
    expand = zone['expand']
    
    offset_grids = []
    
    
    # Offset the grids by km
    if expand:
        for grid in grids:
            pc = pyclipper.PyclipperOffset()    

            pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
            offset_grid = [(point[0], point[1]) for point in pc.Execute(expand *2*1000)[0]]
        
            offset_grids.append(offset_grid)
    else:
        offset_grids = grids
    
    # Union
    pc = pyclipper.Pyclipper()
    pc.AddPaths(offset_grids, pyclipper.PT_SUBJECT, True)
    
    unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
    offset_grids = [[(point[0], point[1]) for point in union] for union in unions]
        
    # Difference with other zones
    other_zone_grids = [z['grids'] for z in aprs_zones_obj if z != zone]
    other_grids = [item for sublist in other_zone_grids for item in sublist]
    
    pc = pyclipper.Pyclipper()
    pc.AddPaths(offset_grids, pyclipper.PT_SUBJECT, True)
    pc.AddPaths(other_grids, pyclipper.PT_CLIP, True)
    
    differences = pc.Execute(pyclipper.CT_DIFFERENCE, pyclipper.PFT_EVENODD, pyclipper.PFT_EVENODD)
    offset_grids = [[(point[0], point[1]) for point in difference] for difference in differences]
    
    zone['grids'] = offset_grids
    zone['simple_grids'] = prepare_simple_grids(zone['grids'])
    zone['simple_outlines'] = fix_antiprime_outlines(ea_grid_to_lonlat(zone['simple_grids']))
    
# Now remove the "default" 144.8 zone
del aprs_zones_obj[-1]
```

#### Let's output these as a header

```python
def file_header_aprs(f):
    print >>f, "/**"
    print >>f, " * Autogenerated aprs zone structures. See sim/geofence"
    print >>f, " */"
    print >>f, ""
    
def header_h_aprs(header, zones):
    print >>header, "#ifndef GEOFENCE_APRS_H"
    print >>header, "#define GEOFENCE_APRS_H"
    print >>header, ""
    print >>header, ""
    print >>header, "#include <stdbool.h>"
    print >>header, ""
    print >>header, "#include \"samd20.h\""
    print >>header, ""
    print >>header, ""
    print >>header, "/**"
    print >>header, " * struct representing all we need to know about a aprs zone"
    print >>header, " */"
    print >>header, "struct aprs_zone_t {"
    print >>header, "  const int32_t** outlines;"
    print >>header, "  const int32_t outline_count;"
    print >>header, "  const uint32_t* outline_lengths;"
    print >>header, "  int32_t frequency, deviation;"
    print >>header, "  char* name;"
    print >>header, "};"
    print >>header, ""
    print >>header, ""
    print >>header, "const struct aprs_zone_t aprs_zones[{0}];".format(len(zones))
    print >>header, ""
    print >>header, "#endif /* GEOFENCE_APRS_H */"
```

```python
def print_outline_aprs(f, zone):
    
    print >>f, u"/* {name:->24} */".format(**zone).encode('utf8')
    
    # Print each array of points
    for o in range(len(zone['simple_outlines'])): # foreach of the countries outlines
    
        outline = zone['simple_outlines'][o]
    
        print >>f, "const int32_t {0}_{1}[] = {{".format(zone['name'].lower() + "_outline", o+1)
        for point in outline:
            # We express locations in µdegrees
            ulon = int(round(point[0]*1e6))
            ulat = int(round(point[1]*1e6))
        
            print >>f, "  {: 10d}, {: 9d},".format(ulon, ulat)
        print >>f, "};"

            
    print >>f, ""
    
    # Print an array of outlines
    print >>f, "const int32_t* {0}_outlines[] = {{".format(zone['name'].lower())
    for o in range(len(zone['simple_outlines'])): # foreach of the countries outlines
         print >>f, "  {0}_{1},".format(zone['name'].lower() + "_outline", o+1)
    print >>f, "};"
    
    # Print an array of their lengths
    print >>f, "const uint32_t {0}_outline_lengths[] = {{".format(zone['name'].lower())
    for outline in zone['simple_outlines']: # foreach of the countries outlines
         print >>f, "  {0},".format(len(outline))
    print >>f, "};"
    print >>f, ""
    print >>f, ""
```

```python
from __future__ import unicode_literals

def print_aprs_zone(f, zone):
    zone['outline_n'] = zone['name'].lower() + "_outlines"
    zone['outline_c'] = len(zone['simple_outlines'])
    zone['outline_l'] = zone['name'].lower() + "_outline_lengths"
    
    print >>f, u"                                              \
    /* {name:->24} */".format(**zone).encode('utf8')
    print >>f, u"  {{ .outlines = {outline_n}, .outline_count = {outline_c}, .outline_lengths = {outline_l},\n\
    .frequency = {frequency}, .deviation = {deviation}, .name = \"{name}\" }},".format(**zone).encode('utf8')
    
def print_aprs_zones(f, zones):
    print >>f, "const struct aprs_zone_t aprs_zones[] = {"
    [print_aprs_zone(f, z) for z in zones]
    print >>f, "};"
```

```python
def write_aprs_zones():
    # header
    header = open('../../firmware/inc/geofence_aprs.h','w')

    file_header_aprs(header)
    header_h_aprs(header, aprs_zones_obj)

    # source
    source = open('../../firmware/src/geofence_aprs.c', 'w')

    file_header(source)
    print >>source, ""
    print >>source, "#include <stdbool.h>"
    print >>source, ""
    print >>source, "#include \"samd20.h\""
    print >>source, "#include \"geofence_aprs.h\""
    print >>source, ""
    print >>source, ""
    print >>source, "/* Longitude,  Latitude */"
    print >>source
    [print_outline_aprs(source, z) for z in aprs_zones_obj]
    print_aprs_zones(source, aprs_zones_obj)

    header.close()
    source.close()
```

```python
write_aprs_zones()
```
