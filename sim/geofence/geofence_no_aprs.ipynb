## Union

```python
# Collect shifted grids
no_aprs_grids_collection = [c['shifted_grids'] for c in countries if c['isocode'] in no_aprs]
no_aprs_grids = [grid for grids in no_aprs_grids_collection for grid in grids]
```

```python
# Union
pc = pyclipper.Pyclipper()
pc.AddPaths(no_aprs_grids, pyclipper.PT_SUBJECT, True)
    
unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
no_aprs_shifted_grids = [[(point[0], point[1]) for point in union] for union in unions]
```

```python
def prepare_simple_grids(grids, tolerance=2*12*1000): # Approx 12km
    simple = [simplify_points(points, tolerance) for points in grids]
    return [s for s in simple if len(s) > 2]
```

```python
# Simplify. 8cos(lat) km
no_aprs_simple_grids = prepare_simple_grids(no_aprs_shifted_grids, tolerance=2*4*1000) 
```

```python
# Back to outlines
no_aprs_simple_outlines = fix_antiprime_outlines(ea_grid_to_lonlat(no_aprs_simple_grids))
```

```python
# Filter out zero-length outlines
filtered_outlines = []
for outline in no_aprs_simple_outlines:
    if (len(outline) > 0):
        filtered_outlines.append(outline)
            
no_aprs_simple_outlines = filtered_outlines
```

```python
# Map
m = Basemap(projection='mill')
 
m.fillcontinents(color='0.96')
m.drawcoastlines()

draw_outlines(m, no_aprs_simple_outlines, '#ff0000')
```
