## Expand and Union

```python
# Generate grids
telemetry_grids = lonlat_to_ea_grid(telemetry_simple_outlines)
```

```python
# Now expand and union
telemetry_shifted_grids = []

# Offset the grids by 450cos(lat) km
for grid in telemetry_grids:
    pc = pyclipper.PyclipperOffset()    
        
    if grid:
        pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
        shifted_grid = [(point[0], point[1]) for point in pc.Execute(450 *2*1000)[0]]
        
        telemetry_shifted_grids.append(shifted_grid)    
    
# Union
pc = pyclipper.Pyclipper()
pc.AddPaths(telemetry_shifted_grids, pyclipper.PT_SUBJECT, True)
    
unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
telemetry_shifted_grids = [[(point[0], point[1]) for point in union] for union in unions]
```

```python
def prepare_simple_grids(grids, tolerance=2*12*1000): # Approx 12km
    simple = [simplify_points(points, tolerance) for points in grids]
    return [s for s in simple if len(s) > 2]
```

```python
# Simplify
telemetry_simple_grids = prepare_simple_grids(telemetry_shifted_grids, tolerance=20 * 2000) # ~20km
```

```python
# Back to outlines
telemetry_shifted_outlines = ea_grid_to_lonlat(telemetry_simple_grids)
```

```python
# Map
m = Basemap(projection='mill')
 
m.fillcontinents(color='0.96')
m.drawcoastlines()

draw_outlines(m, telemetry_shifted_outlines, '#ff0000')
```
