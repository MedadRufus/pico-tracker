## Expand and Union

```python
# Generate grids
lrange_telemetry_grids = lonlat_to_ea_grid(lrange_telemetry_simple_outlines)
additional_telemetry_grids = lonlat_to_ea_grid(additional_telemetry_simple_outlines)
```

```python
# Now expand and union
telemetry_shifted_grids = []

# Offset the grids by 500 km for telemetry_required areas
for grid in lrange_telemetry_grids:
    pc = pyclipper.PyclipperOffset()    
        
    if grid:
        pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
        shifted_grid = [(point[0], point[1]) for point in pc.Execute(500*1000)[0]]
        
        telemetry_shifted_grids.append(shifted_grid)    
        
# Offset the grids by 50 km for additional telemetry areas
for grid in additional_telemetry_grids:
    pc = pyclipper.PyclipperOffset()    
        
    if grid:
        pc.AddPath(grid, pyclipper.JT_ROUND, pyclipper.ET_OPENROUND)
        shifted_grid = [(point[0], point[1]) for point in pc.Execute(50*1000)[0]]
        
        telemetry_shifted_grids.append(shifted_grid)  
    
# Union
pc = pyclipper.Pyclipper()
pc.AddPaths(telemetry_shifted_grids, pyclipper.PT_SUBJECT, True)
    
unions = pc.Execute(pyclipper.CT_UNION, pyclipper.PFT_POSITIVE, pyclipper.PFT_POSITIVE)
telemetry_shifted_grids = [[(point[0], point[1]) for point in union] for union in unions]
```

```python
def prepare_simple_grids(grids, tolerance=10*1000): # 10km tolerance
    simple = [simplify_points(points, tolerance) for points in grids]
    return [s for s in simple if len(s) > 2]
```

```python
# Simplify
telemetry_simple_grids = prepare_simple_grids(telemetry_shifted_grids, tolerance=20*1000) # 20km tolerance
```

```python
# Back to outlines
telemetry_shifted_outlines = fix_antiprime_outlines(ea_grid_to_lonlat(telemetry_simple_grids))
```
